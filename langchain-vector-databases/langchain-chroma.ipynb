```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from langchain_core.documents import Document

doc1 = Document(
    page_content="Virat Kohli is one of India's most successful batsmen and a former captain of the national team. He is known for his consistency across all formats and his aggressive batting style.",
    metadata={"name": "Virat Kohli", "role": "Batsman", "state": "Delhi"}
)

doc2 = Document(
    page_content="Rohit Sharma is the captain of the Indian cricket team and an opening batsman. He holds the record for the highest individual score in ODI cricket.",
    metadata={"name": "Rohit Sharma", "role": "Batsman", "state": "Maharashtra"}
)

doc3 = Document(
    page_content="MS Dhoni is a legendary wicketkeeper-batsman who led India to multiple ICC trophies. He is famous for his calm leadership and finishing abilities in limited-overs cricket.",
    metadata={"name": "MS Dhoni", "role": "Wicketkeeper", "state": "Jharkhand"}
)

doc4 = Document(
    page_content="Jasprit Bumrah is one of the world's premier fast bowlers and a key player for India. He is known for his unique bowling action and deadly yorkers.",
    metadata={"name": "Jasprit Bumrah", "role": "Fast Bowler", "state": "Gujarat"}
)

doc5 = Document(
    page_content="Ravindra Jadeja is a world-class all-rounder contributing with both bat and ball. He is also regarded as one of the best fielders in international cricket.",
    metadata={"name": "Ravindra Jadeja", "role": "All-Rounder", "state": "Gujarat"}
)
```

```python
docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
vector_store = Chroma(
    collection_name="sample",
    embedding_function=embeddings,
    persist_directory="my_new_chroma_db"
)
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.similarity_search(
    query='Who does Wicketkeeper among all the players team?',
    k=2
)
```
